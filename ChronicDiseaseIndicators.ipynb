The goal of this project is to analyzie chronic disease indicators for the US population form years 2001 to 2016. Owing to the large number of values in the dataset, we are limiting our analysis for diseases such as 'Arthritis','Asthma', 'Cancer', 'Cardiovascular Disease', 'Chronic Kidney Disease', 'Chronic Obstructive Pulmonary Disease' and 'Diabetes'.

We are also going to try to predict the prevalence of these chronic disease using indicators such as alcohol use, smoking, high school completion, healthy eating habits, excercise, weight and mental health.

The chronic indicator data is distributed by the state and stratified by gender and ethnicity. 

```python
import pandas as pd
import numpy as np
```

```python
import os
os.path.isfile('/Users/Diwan/U.S._Chronic_Disease_Indicators.csv')
```

```python
# taking a first look at our raw data
CDI = pd.read_csv("U.S._Chronic_Disease_Indicators.csv", low_memory=False)
CDI.head(30)
```

```python
# only selecting columns that we need for analysis

cols = ['YearStart', 'YearEnd', 'State', 'Chronic Disease', 'Indicator', 'DataValueUnit', 'DataValueType', 'DataValue', 'StratificationCategory1', 'Stratification1' ]
CDI = CDI[cols]
CDI.head()
```

```python
#dropping overall stratification category to make our data gender specific

CDI.drop(CDI.loc[CDI['StratificationCategory1']== 'Overall'].index, inplace=True)
CDI.head()
```

```python
# grouping the data by chronic diseases to find relationship between indicators and a specific disease

CDI_grouped= CDI.groupby('Chronic Disease')
CDI_grouped.head()
```

One interesting thing to note in this dataset is that when we group the data by chronic diseases, it seems like certain indicators are only associated with a specific chronic condition.

Another interesting thing to note is that Data Value Unit and Data Value type is also specific to certain indicators so we would not have to further classify the data based on these.
